```python
import numpy as np
np.random.seed(1)

true_p=0.7
data=np.random.binomial(1,true_p,10)

print("observed coin tosses(1=Head,0=Tail)")
```

```python
def log_likelihood(p,data):
  return np.sum(data*np.log(p)+(1-data)*np.log(1-p))
```

```python
import matplotlib.pyplot as plt
p_values=np.linspace(0.01,0.99,100)
ll_values=[log_likelihood(p,data) for p in p_values]
plt.plot(p_values,ll_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

true_p=0.7
sample_sizes=[10,30,50,100,500,1000]
p_mle_values = [] # Initialize p_mle_values as an empty list

for n in sample_sizes:
  data=np.random.binomial(1,true_p,n)
  p_mle=data.mean()
  p_mle_values.append(p_mle)

#plot
plt.plot(sample_sizes,p_mle_values,marker='o')
plt.axhline(true_p,color='r',linestyle='--')
plt.xlabel('Sample Size')
plt.ylabel('MLE of p')
plt.title('MLE convergence with increasing data')
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

true_mu = 35        # true mean
true_sigma = 15    # true standard deviation

sample_sizes = [10, 30, 50, 100, 500, 1000]
mu_mle_values = []

for n in sample_sizes:
    data = np.random.normal(true_mu, true_sigma, n)
    mu_mle = data.mean()   # MLE of mean
    mu_mle_values.append(mu_mle)

# plot
plt.plot(sample_sizes, mu_mle_values, marker='o')
plt.axhline(true_mu, color='r', linestyle='--')
plt.xlabel('Sample Size')
plt.ylabel('MLE of ')
plt.title('MLE convergence for Normal Distribution')
plt.show()
```
